## The Optimizer in Machine Learning

### Momentum

通过记忆之前的梯度, 来平滑当前的梯度(去噪声), 从而得到更好的下降方向. 
![momentum](https://docs.google.com/uc?id=15SJ81VNiDB2KNGXCgTYk0U5S7JPOOHa3)


```python
weight.avg = beta * weight.avg + (1-beta) * weight.grad
new_weight = weight - lr * weight.avg
```

### RMSProp

在全局学习率的基础上, 给没一个参数一个自适应的学习率. 该学习率可以由其梯度的情况(若梯度很小则可以用大学习率, 反之则需要用小的学习率)进行自适应地调节.

instead of using the same learning rate for every parameter, each parameter gets its own specific learning rate controlled by a global learning rate. 

```python
w.square_avg = alpha * w.square_avg + (1-alpha) * (w.grad ** 2)
new_w = w - lr * w.grad / math.sqrt(w.square_avg + eps)
```

### Adam

Adam结合了momentum和RMSProp两种策略, 并使用unbias moving average.

```python
w.avg = beta1 * w.avg + (1-beta1) * w.grad
unbias_avg = w.avg / (1 - (beta1**(i+1)))
w.sqr_avg = beta2 * w.sqr_avg + (1-beta2) * (w.grad ** 2)
new_w = w - lr * unbias_avg / sqrt(w.sqr_avg + eps)
```
